```python
import duckdb
from skimpy import skim
```

```python
# Création d'une BD virtuelle
conn = duckdb.connect()
```

## Fichier des clients

```python
# Récupération de toutes les données du fichier .csv en DF
customers_df = conn.execute(
    """
    SELECT *
    FROM read_csv_auto('data/indexE/Clients.csv', header=True)
    """
).df()
```

```python
# Données stats de la DF
skim(customers_df)
```

```python
# Enregistrement de la DF ci-avant dans une nouvelle BD virtuelle (SQL : CTE)
conn.register('customers_df', customers_df)
```

## Requêtes SQL

Ensemble des clients dont le nom commence par la lettre 'C'

```python
conn.execute(
    """
    -- Récupération des champs
    SELECT *
    FROM customers_df
    -- Filtre
    WHERE Nom LIKE 'C%'
    -- Trie
    ORDER BY Nom ASC
    """
).df()
```

Ensemble des clients dont le nom commence par la lettre 'C' et qui finit par la lettre 'O'

```python
conn.execute(
    """
    -- Champs à récupérer
    SELECT *
    FROM customers_df
    -- Filtre
    WHERE Nom LIKE 'C%o'
    -- Trie
    ORDER BY Nom
    """
).df()
```

Ensemble des clients dont le nom commence par la lettre 'C' et que le prénom se finit par 'Y'

```python
conn.execute(
    """
    -- Champs à récupérer
    SELECT *
    FROM customers_df
    -- Filtre
    WHERE Nom LIKE 'C%' AND Prenom LIKE '%y'
    -- Trie
    ORDER BY Nom ASC
    """
).df()
```

Donner les noms des clients qui contiennent la lettre 'N'


```python
conn.execute(
    """
    -- Champs à récupérer
    SELECT *
    FROM customers_df
    -- Filtre
    WHERE Nom LIKE '%n%'
    -- Trie
    ORDER BY Nom
    """
).df()
```

Donner les prénoms de clients qui contiennent les lettres 'ATH'

```python
conn.execute(
    """
    -- Champs à récupérer
    SELECT *
    FROM customers_df
    -- Filtre
    WHERE Prenom LIKE '%ath%'
    -- Trie
    ORDER BY Nom ASC
    """
).df()
```
