Même fonctionnalité que LIKE mais l'instruction est plus simple et les expressions
régulières sont également utilisées en pandas / polars

> Syntaxe :
> ```
> SELECT nom_colonne
> FROM nom_table
> WHERE nom_colonne REGEXP 'pattern';
> ```
> <br>

````python
import duckdb
from duckdb.typing import BOOLEAN, VARCHAR
from skimpy import skim
import re
````

````python
# Fonction pour effectuer des requêtes SQL avec du REGEX
def regexp(pattern: str, string: str) -> bool:
    return bool(re.search(pattern, string))
````

````python
# Connexion à la librairie duckdb
conn = duckdb.connect()
````

````python
# Création de la fonction ci-avant pour les requêtes SQL
conn.create_function("regexp", regexp, [VARCHAR, VARCHAR], BOOLEAN)
````

## Fichier des produits

````python
# Requête SQL : récupération de toutes les données du fichier .csv en DF
products_df = conn.execute(
    """
    SELECT *
    FROM read_csv_auto('data/indexE/Produits.csv', header=True);
    """
).df()
````

````python
# Stats de la DF
skim(products_df)
````

````python
# Enregistrement de la DF ci-avant dans une BD virtuelle (requête SQL : VIEW)
conn.register('products_db', products_df)
````

## Fichier des clients

````python
# Requête SQL : récupération de toutes les données du fichier .csv en DF
customers_df = conn.execute(
    """
    SELECT *
    FROM read_csv_auto('data/indexE/Clients.csv', header=True);
    """
).df()
````

````python
# Données stats de la DF
skim(customers_df)
````

````python
# Enregistrement de la DF ci-avant en BD virtuelle (SQL : requête VIEW)
conn.register('customers_db', customers_df)
````

## Requêtes SQL

Liste des produits qui contiennent 'TV' avec LIKE

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre
    WHERE Description LIKE '%TV%';
    """
).df()
````

Liste des produits qui contiennent 'TV' avec REGEXP

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre avec pour argument : pattern, colonne ciblée
    WHERE regexp('TV', NomProduit);
    """
).df()
````

Liste des produits dont le nom commence par la lettre D avec LIKE

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre
    WHERE NomProduit LIKE 'D%'
    -- Trie
    ORDER BY NomProduit ASC;
    """
).df()
````

Liste des produits dont le nom commence par la lettre D avec REGEXP

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre avec en argument : pattern, nom de la colonne ciblée
    WHERE regexp('^D', NomProduit)
    -- Trie
    ORDER BY NomProduit ASC;
    """
).df()
````

Liste des produits qui se termine par la lettre M avec LIKE

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre
    WHERE NomProduit LIKE '%m';
    """
).df()
````

Liste des produits qui se termine par la lettre M avec REGEXP

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT *
    -- BD récupérée
    FROM products_db
    -- Filtre avec pour arguments : pattern, nom de la colonne ciblée
    WHERE regexp('m$', NomProduit);
    """
).df()
````
