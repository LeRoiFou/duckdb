Avec l'instruction JOIN, les identifiants sont communs aux deux tables
L'instruction ON affiche la colonne commune des deux tables à deux reprises si on demande d'afficher toutes les colonnes des tables à la différence de l'instruction USING() -> dans ce cas, la colonne commune des deux tables n'apparaît qu'une seule fois.

> Syntaxe :
> ```
> SELECT 
>     t1.nom_colonne
>     , t2.nom_colonne
> FROM table1 t1
> JOIN table2 t2
> ON t1.cle_colonne = t2.cle_colonne;
> ```

## Requêtes SQL

Donner pour chaque vente, le nom et le prénom de l'employé qui réalisé la vente

````python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        s.VenteID AS "Numéro de la vente"
        , e.EmployeID
        , e.Nom
        , e.Prenom
        , s.MontantTotal AS "Montant total"
    -- BD récupérées
    FROM employees_db e
    JOIN sales_db s
    ON e.EmployeID = s.EmployeID
    -- Trie
    ORDER BY "Numéro de la vente" ASC;
    """
).df()
````

Donner pour chaque produit de la base de données le nom, l'adresse et le n° de téléphone de son fournisseur

````python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        products.NomProduit AS "Nom du produit"
        , suppliers.NomFournisseur AS "Nom du fournisseur"
        , suppliers.Adresse
        , suppliers.NumeroTelephone AS Telephone
    -- BD récupérées
    FROM products_db products
    JOIN suppliers_db suppliers
    ON products.FournisseurID = suppliers.FournisseurID
    -- Trie
    ORDER BY "Nom du produit" ASC;
    """
).df()
````

Donner le nom et prénom des employés ayant réalisé la somme des ventes les plus élevées

````python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        e.Nom
        , e.Prenom
        , ROUND(SUM(s.MontantTotal), 0) AS "Total des ventes"
    -- BD récupérées
    FROM employees_db e
    JOIN sales_db s
    ON e.EmployeID = s.EmployeID
    -- TCD
    GROUP BY 
        e.Nom
        , e.Prenom
    -- Trie décroissant
    ORDER BY "Total des ventes" DESC;
    """
).df()
````

Donner pour chaque client, le nom, le prénom, l'adresse ainsi que le nombre d'achats réalisé

````python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        c.Nom
        , c.Prenom
        , c.Adresse
        , COUNT(s.VenteID) AS "Nombre d'achats"
    -- BD récupérées
    FROM customers_db c
        JOIN sales_db s
        ON c.ClientID = s.ClientID
    -- TCD
    GROUP BY
        c.Nom
        , c.Prenom
        , c.Adresse
    -- Trie
    ORDER BY 
        c.Nom ASC
        , c.Prenom ASC;
    """
).df()
````
