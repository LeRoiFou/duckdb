<b> Objet de la requête : </b><br>
Une clause JOIN est utilisée pour combiner des lignes provenant de deux tables ou plus, sur la base d'une colonne liée entre elles.

Voici les différents types de JOIN en SQL :
- (INNER) JOIN : renvoie les enregistrements dont les valeurs correspondent dans les deux tables.
- LEFT (OUTER) JOIN : renvoie tous les enregistrements de la table de gauche et les enregistrements correspondants de la table de droite.
- RIGHT (OUTER) JOIN : renvoie tous les enregistrements de la table de droite et les enregistrements correspondants de la table de gauche.
- FULL (OUTER) JOIN : renvoie tous les enregistrements lorsqu'il existe une correspondance dans la table de gauche ou de droite.
