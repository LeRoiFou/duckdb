L'instruction VIEW permet de créer une "table" enregistrée en tant que "variable" qui prend moins de mémoire que de générer une table et permet d'être récupérer pour effectuer d'autres requêtes SQL dessus.
La vue créée se trouve dans le menu à gauche "SCHEMAS" (rafraîchir la base de données)


```python
import duckdb
from skimpy import skim
```

```python
# Connection à la librairie duckdb
conn = duckdb.connect()
```

## Fichier des ventes

```python
# Requête SQL : récupération de toutes les données du fichier .csv en DF
sales_df = conn.execute(
    """
    SELECT *
    FROM read_csv_auto('data/indexE/Ventes.csv', header=True);
    """
).df()
```

```python
# Stats de la DF
skim(sales_df)
```

```python
# Enregistrement de la DF en BD virtuelle (requête SQL : VIEW)
conn.register('sales_db', sales_df)
```

## Requêtes SQL

Créez une BD virtuelle sur les ventes de l'année 2021

```python
sales2021_df = conn.execute(
    """
    -- Récupération des colonnes
    SELECT *
    -- BD récupérée
    FROM sales_db
    -- Filtre
    WHERE YEAR(DateVente) = '2021';
    """
).df()
```

```python
# Stats de la DF
skim(sales2021_df)
```

```python
# Enregistrement de la DF dans une BD virtuelle (requête SQL : VIEW)
conn.register('sales2021_db', sales2021_df)
```

À partir de cette BD virtuelle créée, donner la liste des clients qui ont réalisé plus d'un achat en 2021

```python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        c.ClientID
        , c.Nom
        , c.Prenom
        , COUNT(s21.VenteID) AS "Nombre d'achats"
    -- BD récupérées
    FROM customers_db c
    JOIN sales2021_db s21
    ON c.ClientID = s21.ClientID
    -- TCD
    GROUP BY
        c.ClientID
        , c.Nom
        , c.Prenom
    -- Filtre
    HAVING "Nombre d'achats" > 1
    -- Trie
    ORDER BY "Nombre d'achats" DESC;
    """
).df()
```

À partir de la BD virtuelle créée, donner la liste des employés qui ont réalisé des ventes moyennes supérieures à 500 € en 2021

```python
conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        e.EmployeID
        , e.Nom
        , e.Prenom
        , ROUND(AVG(s21.MontantTotal), 0) AS "Moyenne des ventes réalisées"
    -- BD récupérées
    FROM employees_db e
    JOIN sales2021_db s21
    ON e.EmployeID = s21.EmployeID
    -- TCD
    GROUP BY 
        e.EmployeID
        , e.Nom
        , e.Prenom
    -- Filtre
    HAVING "Moyenne des ventes réalisées" > 500
    -- Trie
    ORDER BY
        "Moyenne des ventes réalisées" DESC
        , e.Nom ASC;
    """
).df()
```

Créer une BD virtuelle contenant les produits dont le prix est > à 500 €

```python
topprice_df = conn.execute(
    """
    -- Colonnes à récupérer
    SELECT
        p.ProduitID
        , p.NomProduit AS "Nom du produit"
        , p.Description
        , p.PrixUnitaire AS "Prix unitaire"
    -- BD récupérée
    FROM products_db p
    -- Filtre
    WHERE "Prix unitaire" > 500;
    """
).df()
```

```python
# Stats de la DF
skim(topprice_df)
```

```python
# Enregistrement de la DF ci-avant dans une BD virtuelle (requête SQL : VIEW)
conn.register('topprice_db', topprice_df)
```

```python
# Affichage des données de la BD virtuelle créée ci-avant
conn.execute(
    """
    SELECT *
    FROM topprice_db;
    """
).df()
```
